While Loop

```python
age = 15
while age < 18:
    print('student',age)
    age = age + 1
```

```python
time = 0
while time <= 10:
    print(time)
    time = time + 1
```

```python
button = 'hello'
while button != 'exit':
    print(button)
    button = input("eater 'exit' to stop loop")
```
